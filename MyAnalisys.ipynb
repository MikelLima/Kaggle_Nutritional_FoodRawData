## Create a function that selects the 10 foods that contain the union of highest of Nutrient1 and highest of Nutrient2

Create a function that identifies which food names are found in both the Rankings1 and Rankings2 at the same time.
Then, rank the foods found in both Rankings based on the highest compatibility. 

```python
def unionOfParameters(Nutrient1:str, Nutrient2:str, NumberOfChecks:int) -> tuple[int,list[str]]:
    '''
    This functions checks if there is one or multiple foods that are contained in 2 of the rankings created above. 
    It outputs a list of the food names that are contained in both Rankings
    If there is an error, an error int will be outputed to explain the reason. 
        0: No Error. The list of FoodNames will be oredered by compatibility.
        1: Parameter <Nutrient1> is is not a column of the Rankings DataFrame
        2: Parameter <Nutrient2> is is not a column of the Rankings DataFrame
        3: Parameter <NumberOfChecks> is not in an acceptable range.
        4: No foods were found for those Nutrients in the DataFrame columns 
        5: No foods were found that were both in the Rankings1 and Rankings2 at the same time. 
    '''
    _Error:int = 0

    if (Nutrient1 not in Data.columns):
        print(f"{Nutrient1} is not a Dictionary Entry of the DataFrame.")
        _Error=1
        return (_Error,[])
    if (Nutrient2 not in Data.columns):
        print(f"{Nutrient2} is not a Dictionary Entry of the DataFrame.")
        _Error=2
        return (_Error,[])
    if (NumberOfChecks<1) or (NumberOfChecks>len(inputData)):
        _Error=3
        return (_Error,[])
    if (Nutrient1 == "food" or Nutrient1 == "food_normalized") or (Nutrient2 == "food" or Nutrient2 == "food_normalized"):
        _Error=4
        return (_Error,[])        

    _Rankings1 = Rankings[Nutrient1]
    _Rankings2 = Rankings[Nutrient2]

    # Select how many names of foods are to be compared. 
    _Foods1 = _Rankings1.food.head(NumberOfChecks)
    _Foods2 = _Rankings2.food.head(NumberOfChecks)

    _isInBoth:list[str] = [food1 for food1, food2 in zip(_Foods1, _Foods2) if food1 == food2]
    # for food1 in _Foods1:
    #     for food2 in _Foods2:
    #         if (food1 == food2): 
    #             _isInBoth.append(food1)

    # If no food name is in both rankings of foods...
    if len(_isInBoth)==0:
        _Error=5
        return (_Error,[])

    return (_Error, _isInBoth)

# ErrorValue,ListOfFoods = unionOfParameters("Iron","Protein",3400)   
# print(ErrorValue,ListOfFoods)
```

Create a function that ranks the foods found in both Rankings based on the highest compatibility. 

```python
def rankUnionOfFoods(Nutrient1:str, Nutrient2:str, foodList:list[str], ascending:bool=False) -> list[str]:
    '''
    This function ranks a list of foods based on their compatibility with two specified nutrients.
    Compatibility is calculated using the normalized values of the two nutrients for each food.
    The function returns the list of food names sorted by their compatibility in ascending or descending order.
    '''
    import math
    
    _Rankings1:pd.DataFrame = Rankings[Nutrient1]
    _Rankings2:pd.DataFrame = Rankings[Nutrient2]

    _CompatibilityVariables:dict = {}
    _Compatibility:dict = {}

    # _ReducedRankings:pd.DataFrame 
    for idx,foodName in enumerate(foodList):
        # Obtain the individual values of Nutrient1 and Nutrient2 from each food, and the max value of the DataFrame for that Nutrient
        _a = float(_Rankings1[_Rankings1.food==foodName][Nutrient1])
        _a_max = float(_Rankings1[Nutrient1].max())

        _b = float(_Rankings2[_Rankings2.food==foodName][Nutrient2])
        _b_max = float(_Rankings1[Nutrient2].max())

        # Obtain the Normalized compatibility value
        _CompatibilityVariables[foodName] =(_a,_b)
        _Compatibility[foodName]=round(math.sqrt((_a/_a_max*100)**2 + (_b/_b_max*100)**2),3)

    # Reorder the food names in descending order (the ones with higher comatibility go first)
    # Sort the dictionary by values and return the keys in the desired order
    sorted_foods = sorted(_Compatibility, key=_Compatibility.get, reverse=not ascending)
    return sorted_foods


# rankUnionOfFoods("Iron","Protein",ListOfFoods)
```

```python
dictOfFoods:dict = {}

for nutrient1 in Data.columns:
    for nutrient2 in Data.columns:
        if (nutrient1 != nutrient2):
            _err, listOfFoods = unionOfParameters(nutrient1,nutrient2,3400)
            if _err == 0:
                listOfFoods=rankUnionOfFoods(nutrient1,nutrient2,listOfFoods)
                dictOfFoods[(nutrient1,nutrient2)] = listOfFoods
                print(nutrient1,"   + ",nutrient2," -> ",dictOfFoods[(nutrient1,nutrient2)])

print(len(dictOfFoods))
```
