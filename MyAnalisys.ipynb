# Objective
To analyse which foods are richer in each of the micro and macro-nutrients, and the different relationships between them.

```python
import csv
import pandas as pd
import matplotlib.pyplot as plt
```

## Obtain the data from the appropriate .csv file
Data was obtained from: https://www.kaggle.com/datasets/tharunmss/nutritional-breakdown-of-foods/data


```python
inputData:list = []
with open("cleaned_nutrition_dataset.csv", "r", newline="\n") as my_csv_file:
    my_csv_file_fieldnames:list[str] = ["Vitamin C","Vitamin B11","Sodium","Calcium","Carbohydrates","food","Iron","Caloric Value","Sugars","Dietary Fiber","Fat","Protein","food_normalized"]
    csv_reader = csv.DictReader(f=my_csv_file, delimiter=",") # , fieldnames=my_csv_file_fieldnames,

    for row in csv_reader:
        inputData.append(row)

print(inputData[0:5])
```

```python
# Show the dictionary to check its READ was ok
print(inputData[0])
# inputData.remove[0,:]

# Check its length
print(f"The length of this inputData is {len(inputData)}")
```

## Ready the data into a Pandas Structure
Use the function pd.DataFrame to change the Dictionary into a Pandas DataFrame

```python
Data=pd.DataFrame(inputData)
Data.head()
```

## Rank the foods in order of Vitamin C contents

```python
VitaminC_Ranking = Data.sort_values("Vitamin C", ascending=False)
VitaminC_Ranking.head(10)
```
